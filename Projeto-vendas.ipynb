```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)

import pandas as pd

sales = pd.read_csv('sales_data_sample.csv',encoding='latin1')

sales.head()

sales = sales.drop(['ADDRESSLINE2','POSTALCODE','CONTACTLASTNAME','QTR_ID','MONTH_ID','YEAR_ID'], axis=1)

sales
```

```python
sales.isnull().sum()


#state e territory tem valores nulos

sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
sales['ORDERDATE_BR'] = sales['ORDERDATE'].dt.strftime('%d/%m/%Y')

sales
sales = sales.drop(columns=['ORDERDATE_BR'])
```

```python
sales
```

# ðŸŽ¯ **Desafio 10 â€” AnÃ¡lise combinada (nÃ­vel intermediÃ¡rio â†’ avanÃ§ado)**

**Objetivo:** descobrir os **5 produtos mais valiosos** considerando **faturamento e quantidade vendida juntos**.


```python
vendas_total = sales.groupby('PRODUCTCODE')[['QUANTITYORDERED','SALES']].sum().sort_values(by='SALES', ascending=False).head(5)
vendas_total
```

# ðŸŽ¯ **Desafio 11 â€” PreÃ§o unitÃ¡rio e anÃ¡lise avanÃ§ada**

**Objetivo:** criar uma tabela com os **5 produtos mais valiosos**, incluindo:



```python
faturamento = sales.groupby('PRODUCTCODE')[['QUANTITYORDERED','SALES']].sum().sort_values(by='SALES',ascending=False).head(5)

faturamento['UNITARIO'] = faturamento['SALES']/faturamento['QUANTITYORDERED']

faturamento
```

# ðŸŽ¯ **Desafio 12**

Descubra como o faturamento evoluiu ao longo do tempo usando a coluna `ORDERDATE`.
Mostre o faturamento total por ano.


```python
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'], errors='coerce')
tempo = sales[['ORDERDATE','PRODUCTCODE','SALES']].copy()

tempo.loc[:, 'ANO'] = tempo['ORDERDATE'].dt.year

df_ano = tempo.groupby('ANO')['SALES'].sum()

df_ordenado = df_ano.sort_values(ascending=False)

display(df_ordenado)
```

```python
df_ordenado = tempo.sort_values(by='SALES', ascending=False)

df_ordenado
```
